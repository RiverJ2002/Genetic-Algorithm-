```python
import numpy as np
from scipy.spatial.distance import cdist
import pandas as pd
import random
```

```python
city_pos = [[60, 200], [180, 200], [80, 180], [140, 180], [20, 160], [100, 160], [200, 160], [140, 140], [40, 
120], [100, 120], [180, 100], [60, 80], [120, 80], [180, 60], [20, 40], [100, 40], [200, 40], [20, 
20], [60, 20], [160, 20]]
print("Number of cities(genes):",len(city_pos))
```

```python
distances = cdist(city_pos, city_pos)
distances.shape
```

```python
#what can I do? life is beautiful but you don't have clue. black beauty

class GA:
    def __init__(self,city_pos , distances,population_size,num_generations,mutation_rate,tournament_size,waiting_callback):
        
        self.city_pos = city_pos
        self.distances = distances
        self.population_size = population_size
        self.num_generations = num_generations
        self.mutation_rate = mutation_rate
        self.tournament_size = tournament_size
        self.waiting_callback = waiting_callback
        
        
    def fitness(self,chromosome):
        total_distance = 0
        for i in range(len(chromosome)-1):
            current_city = chromosome[i]
            next_city = chromosome[i + 1]
            total_distance += self.distances[current_city][next_city]

        return total_distance

    def generate_population(self,population_size):
        population = []
        for _ in range(population_size):
            chromosome = list(range(len(self.city_pos)))
            random.shuffle(chromosome)

            chromosome.append(chromosome[0])

            population.append(chromosome)

        return population

    
    def selection(self, population):
        tournament = random.sample(population, self.tournament_size)
        fittest = min(tournament, key=lambda x: self.fitness(x))
        return fittest


    def crossover(self, parent1, parent2):
        crossover_point = random.randint(1, 19)
        child1 = parent1[:crossover_point] + [city for city in parent2 if city not in parent1[:crossover_point]]
        child2 = parent2[:crossover_point] + [city for city in parent1 if city not in parent2[:crossover_point]]

        return child1, child2


    def mutation(self, chromosome):
        mutated_chromosome = chromosome.copy()

        # Perform mutation on the chromosome
        for _ in range(self.mutation_rate):
            index1 = random.randint(0, len(mutated_chromosome) - 1)
            index2 = random.randint(0, len(mutated_chromosome) - 1)
            mutated_chromosome[index1], mutated_chromosome[index2] = (
              mutated_chromosome[index2],
              mutated_chromosome[index1],
            )

        return mutated_chromosome


    def call_back(self,best_fitnesses,population):
        check_fitness = []
        for i in best_fitnesses[len(best_fitnesses)-self.waiting_callback:]:
            for j in best_fitnesses[len(best_fitnesses)-self.waiting_callback:]:
                check_fitness.append(abs(i-j))
                
        if np.array(check_fitness).mean()<1:
            return self.generate_population(self.population_size)
            
        return population




    def ga(self):
        
        population = self.generate_population(self.population_size)


        best_fitnesses = []

        for generation in range(self.num_generations):

            best_fitness = np.min(list(map(self.fitness, population)))
            best_fitnesses.append(best_fitness)
            population = self.call_back(best_fitnesses,population)


            if generation%100==0:
                print(f"Generation {generation} ------------> average fitness {np.mean(list(map(self.fitness, population)))} , Best Chromosome Fitness {best_fitness}")
                
            offspring = []
            for _ in range(self.population_size // 2):
                
                parent1, parent2 = self.selection(population) , self.selection(population)
                child1, child2 = self.crossover(parent1, parent2)
                child1, child2 = self.mutation(child1) , self.mutation(child2)


                offspring.extend([child1, child2])
            
            population = []
            population.extend(offspring)
            



        best_individual = min(population, key=lambda x: self.fitness(x))
        best_distance = min(best_fitnesses)
        
        return best_individual, best_distance , best_fitnesses



    
```

```python
ga = GA(city_pos,distances,500,2000,2,6,5)
shortest_routs = ga.ga()
        
```
